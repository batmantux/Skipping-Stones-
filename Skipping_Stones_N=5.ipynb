```
import numpy as np
import math
class CriticalVelocity:
    """ Class of Critical Velocity """

    def __init__(self, o, M , a,N):
        self.o = o
        self.g = 9.8
        self.M = M
        self.d = 1000
        self.a = a
        self.C = 1
        self.N = N
   
    def u(self):
        return (np.sin(np.radians(self.o)) + np.cos(np.radians(self.o))) / (np.cos(np.radians(self.o)) - np.sin(np.radians(self.o)))

    def L(self):
        return 2*3.14*np.sqrt((2*(self.M)*np.sin(np.radians(self.o))/(self.C*self.d*self.a)))
   
    def Vc(self):
        return np.sqrt(2*(self.u())*(self.g)*(self.L()))
   
    def Vin(self):
        return np.sqrt(((self.Vc())**2)*(self.N))
```

```
my_list=[]
for mass in np.linspace(0,1,11):
    for length in np.linspace(0,1,11):
        for angle in np.arange(0,45,1):
            Coef = CriticalVelocity(o=angle, M=mass,a=length,N=5)
            Velocity=Coef.Vin()
            Velocity_critical=Coef.Vc()
           
            if 0<Velocity<10:
               z=[mass, length, angle, Velocity]
               my_list.append(z)
my_list               
               
              
```
